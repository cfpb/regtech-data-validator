Let's make sure we didn't lose any validations going from a single schema to a phase 1 and phase 2 schema approach. I'm going to verify that the set of validations is the same between the two. 

```python
from phase_validations import phase_1_and_2_validations
from schema import sblar_schema
```

```python
names_from_phase_1_and_phase_2_validations = {}
count_1 = 0

for column, phases in phase_1_and_2_validations.items():
    
    validations = phases["phase_1"] + phases["phase_2"]
    validation_names = [validation.name for validation in validations]
    names_from_phase_1_and_phase_2_validations[column] = validation_names
    count_1 += len(validation_names)
```

```python
names_from_original = {}
count_2 = 0

for column in sblar_schema.columns:
    
    checks = sblar_schema.columns[column].checks
    validation_names = [check.name for check in checks]
    names_from_original[column] = validation_names
    count_2 += len(validation_names)
```
